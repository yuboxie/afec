```python
import os
import pickle
import random
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
test_posts_df = pd.read_csv('../data/test/posts_labeled_32.csv')
test_posts_df
```

```python
sample_indices = np.load('../data/test/posts_indices_sample_idx.npy')
sample_test_posts_df = test_posts_df.iloc[sample_indices]
sample_test_posts_df
```

# Combine Baseline Results

```python
with open('../data/test/result/sample/moel_result.pickle', 'rb') as f:
    moel_result = pickle.load(f)
print(len(moel_result))
```

```python
with open('../data/test/result/sample/mime_result.pickle', 'rb') as f:
    mime_result = pickle.load(f)
print(len(mime_result))
```

```python
with open('../data/test/result/sample/blender_result.pickle', 'rb') as f:
    blender_result = pickle.load(f)
print(len(blender_result))
```

```python
with open('../data/test/result/sample/cem_result.pickle', 'rb') as f:
    cem_result = pickle.load(f)
print(len(cem_result))
```

```python
baselines_result_dict = {}
baselines_result_cols = ['post_id', 'post', 'emotion', 'moel', 'mime', 'cem', 'blender']
for col in baselines_result_cols:
    baselines_result_dict[col] = []
```

```python
for i in tqdm(range(sample_test_posts_df.shape[0])):
    post_id = sample_test_posts_df.iloc[i]['id']
    post = sample_test_posts_df.iloc[i]['text']
    emotion = sample_test_posts_df.iloc[i]['emotion']
    moel = moel_result[i]
    mime = mime_result[i]
    cem = cem_result[i]
    blender = blender_result[i]
    baselines_result_dict['post_id'].append(post_id)
    baselines_result_dict['post'].append(post)
    baselines_result_dict['emotion'].append(emotion)
    baselines_result_dict['moel'].append(moel)
    baselines_result_dict['mime'].append(mime)
    baselines_result_dict['cem'].append(cem)
    baselines_result_dict['blender'].append(blender)
```

```python
baselines_result_df = pd.DataFrame(baselines_result_dict)
baselines_result_df
```

```python
baselines_result_df.to_csv('../data/test/result/sample/baselines_result.csv', index = False)
```

# Generate CSV for Golden Answers

```python
gold_ans_dict = {}
gold_ans_cols = ['post_id', 'post', 'reply_ids', 'replies']
for col in gold_ans_cols:
    gold_ans_dict[col] = []
```

```python
test_replies_df = pd.read_csv('../data/test/replies.csv')
test_replies_df
```

```python
for i in tqdm(range(sample_test_posts_df.shape[0])):
    post_id = sample_test_posts_df.iloc[i]['id']
    post = sample_test_posts_df.iloc[i]['text']
    df = test_replies_df[test_replies_df['parent_id'] == post_id]
    reply_ids = '\n'.join(df['id'].tolist())
    replies = '\n'.join(df['text'].tolist())
    gold_ans_dict['post_id'].append(post_id)
    gold_ans_dict['post'].append(post)
    gold_ans_dict['reply_ids'].append(reply_ids)
    gold_ans_dict['replies'].append(replies)
```

```python
gold_ans_df = pd.DataFrame(gold_ans_dict)
gold_ans_df
```

```python
gold_ans_df.to_csv('../data/test/result/sample/gold_ans.csv', index = False)
```

# Generate CSV for MEED

```python
test_posts_df = pd.read_csv('../data/test/posts_labeled_41.csv')
test_posts_df
```

```python
sample_indices = np.load('../data/test/posts_indices_sample_idx.npy')
sample_test_posts_df = test_posts_df.iloc[sample_indices]
sample_test_posts_df
```

```python
with open('../data/test/result/sample/meed_result.pickle', 'rb') as f:
    meed_result = pickle.load(f)
print(len(meed_result))
```

```python
with open('../scripts/meed2/ebp_labels.txt', 'r') as f:
    emo2id = {}
    id2emo = {}
    for line in f:
        items = line.strip().split(',')
        emo2id[items[0]] = int(items[1])
        id2emo[int(items[1])] = items[0]
print(id2emo)
```

```python
pred_emo_df = pd.read_csv('../scripts/meed2/emo_pred.csv')
pred_emos = [id2emo[i] for i in pred_emo_df['y_pred'].tolist()]
print(pred_emos)
```

```python
baselines_result_dict = {}
baselines_result_cols = ['post_id', 'post', 'emotion', 'pred_emo', 'meed']
for col in baselines_result_cols:
    baselines_result_dict[col] = []
```

```python
for i in tqdm(range(sample_test_posts_df.shape[0])):
    post_id = sample_test_posts_df.iloc[i]['id']
    post = sample_test_posts_df.iloc[i]['text']
    emotion = sample_test_posts_df.iloc[i]['emotion']
    pred_emo = pred_emos[i]
    meed = meed_result[i]
    baselines_result_dict['post_id'].append(post_id)
    baselines_result_dict['post'].append(post)
    baselines_result_dict['emotion'].append(emotion)
    baselines_result_dict['pred_emo'].append(pred_emo)
    baselines_result_dict['meed'].append(meed)
```

```python
baselines_result_df = pd.DataFrame(baselines_result_dict)
baselines_result_df
```

```python
baselines_result_df.to_csv('../data/test/result/sample/meed_result.csv', index = False)
```

# Combine All Results

```python
retrieval_result_df = pd.read_csv('../data/test/result/all/retrieval_result.csv')
retrieval_result_df
```

```python
with open('../data/test/result/all/moel_result.pickle', 'rb') as f:
    moel_result = pickle.load(f)
print(len(moel_result))
print(moel_result[:5])
```

```python
with open('../data/test/result/all/mime_result.pickle', 'rb') as f:
    mime_result = pickle.load(f)
print(len(mime_result))
print(mime_result[:5])
```

```python
with open('../data/test/result/all/cem_result.pickle', 'rb') as f:
    cem_result = pickle.load(f)
print(len(cem_result))
print(cem_result[:5])
```

```python
with open('../data/test/result/all/blender_result.pickle', 'rb') as f:
    blender_result = pickle.load(f)
print(len(blender_result))
print(blender_result[:5])
```

```python
with open('../data/test/result/all/meed_result.pickle', 'rb') as f:
    meed_result = pickle.load(f)
print(len(meed_result))
print(meed_result[:5])
```

```python
all_result_df = retrieval_result_df.copy()
all_result_df['moel'] = moel_result
all_result_df['mime'] = mime_result
all_result_df['cem'] = cem_result
all_result_df['blender'] = blender_result
all_result_df['meed'] = meed_result
all_result_df
```

```python
test_replies_df = pd.read_csv('../data/test/replies.csv')
test_replies_df
```

```python
groundtruth_id = []
groundtruth = []
for i in tqdm(range(all_result_df.shape[0])):
    parent_id = all_result_df.iloc[i]['post_id']
    gt_df = test_replies_df[test_replies_df['parent_id'] == parent_id]
    assert gt_df.shape[0] > 0
    groundtruth_id.append('\n'.join(gt_df['id'].tolist()))
    groundtruth.append('\n'.join(gt_df['text'].tolist()))
all_result_df['groundtruth_id'] = groundtruth_id
all_result_df['groundtruth'] = groundtruth
all_result_df
```

```python
all_result_df.to_csv('../data/test/result/all/all_result.csv', index = False)
```
